```python
import numpy as np
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
```

# SQL Training with Pandas and SQLite

First we must setup our connection to the database with the following command

```python
con = sqlite3.connect("MEJ.db")
```

We import SQL tables using the SQL commands we just learned with the pandas function "read_sql"
    


```python
exposure_indicators = pd.read_sql("SELECT * FROM exposure;", con, index_col = "Tract_ID")
exposure_indicators.head()
```

Now that we have a table we can do some basic data analysis

```python
exposure_indicators.hist("ozone",edgecolor = "black")
```
